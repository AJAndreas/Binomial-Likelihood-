## The Binomial Likelihood 

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def binomial_distribution(tries, successes, p):
    
    factorial = np.math.factorial(tries) / (np.math.factorial(successes) * np.math.factorial(tries - successes))
    
    pdf = factorial * p**successes * (1-p)**(100-successes)
    
    return pdf 
```

```python
# This returns a value for the PMF for all values of parameter, p, in the range 0-1 (which is a probability)
probs = np.linspace(0, 1, 101)
densities = [binomial_distribution(100, 54, p) for p in np.linspace(0, 1, 101)]
```

```python
# This computes the maximum value of the densities from above 
max_val = max(densities)
print(f'The maximum value of the binomial probability mass function for observing 54 heads in 100 coin tosses is {max_val}')
```

```python
# This indexes the probability values we created earlier with the index of the max value from the densities 
MLE = probs[densities.index(max_val)]
print(f"The value of the parameter, p, which maximises the binomial likelihood is {MLE}")
```

```python
# The plot below shows what the PMF looks like
y = binomial_distribution(100, 54, 0.54)
x = 54
plt.plot(densities)
plt.annotate('MLE', xy =(x, y)) 
plt.show()
```
